# Load Packages and Parse 

```python
import time as time
import xml.etree.ElementTree as ET
from collections import defaultdict
import pandas as pd

start = time.perf_counter()
d = defaultdict(list)
#since u r reading from a file,
# root should be root = ET.parse('filename.xml').getroot()
#mine is wrapped in a string hence :
root = ET.parse('UCL1617_M13_2019285_BEN_BES_PosData_Live.xml').getroot()
#required data is in the Frame section
for ent in root.findall('./Match//Frame'):
    #this gets us the timestamp
    Frame = ent.attrib['utc']
    for entry in ent.findall('Objs/Obj'):
        #append the objects to the relevant timestamp
        d[Frame].append(entry.attrib)

df = (pd.concat((pd.DataFrame(value) #create dataframe of the values
                 .assign(Frame=key) #assign keys to the dataframe
                 .filter(['id','Frame','x','y','z']) #keep only required columns
                 for key, value in d.items()),
                axis=0) #concatenate on the columns axis
     )
end = time.perf_counter()
print(end-start)
```

```python
#Load pickle dataframe

data = pd.read_pickle('dummy_2019285.pkl')
data[['id', 'x','y','z']] = data[['id', 'x','y','z']].astype(float)
data.head()
```

```python
#Locate ball coordinates only 

ind=data.loc[(data["id"] ==0)]
ind.head()
```

```python
#Percentage of the game the ball is in a particular half of the game
ter_lh=len(ind.loc[(ind["x"]< 0)])/len(ind)
print("Territory Left Half")
print(ter_lh)
ter_rh=1-ter_lh
print("Territory Right Half")
print(ter_rh)
```

```python
#Percentage of the game the ball is in each quadrant 
ter_a=len(ind.loc[(ind['x'] <= 0) & (ind['y'] >= 0)])/len(ind)
print("Territory top left quadrant")
print(ter_a)
ter_b=len(ind.loc[(ind['x'] >= 0) & (ind['y'] >= 0)])/len(ind)
print("Territory top right quadrant")
print(ter_b)
ter_c=len(ind.loc[(ind['x'] >= 0) & (ind['y'] <= 0)])/len(ind)
print("Territory bottom right quadrant")
print(ter_c)
ter_d=len(ind.loc[(ind['x'] <= 0) & (ind['y'] <= 0)])/len(ind)
print("Territory bottom left quadrant")
print(ter_d)
```
